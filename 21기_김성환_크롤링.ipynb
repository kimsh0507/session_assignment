```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'
res = requests.get(url)

print(res) # 응답 객체
res.text # HTML 텍스트
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
movie = soup.select('td.title div.tit3')
```

```python
def star(L : list) :
  a = str(L[1])
  for i in range(2,len(L)):
    if L[i] not in ['[',']',',',' '] :
      a += str(L[i])
  
  return a
```

```python
movies_info_list = list()

for movie in movie[:10]:
    movie_link = 'https://movie.naver.com' + movie.select_one('a')['href'] # 태그 내 속성 접근
    response_movie = requests.get(movie_link) # 배우 상세 페이지 이동
    soup_movie = BeautifulSoup(response_movie.text, 'html.parser')
    all_info = soup_movie.select_one('#content > div.article > div.mv_info_area > div.mv_info')

    title = list()
    title_info = all_info.select_one('h3 > a').text
    title.append(title_info)

    main_star_info = soup_movie.select_one('span.st_off').text
    spc_star_info = all_info.select('div.main_score > div > div > a > div > em ')
    ntz_star_info = all_info.select('div.main_score > div > div > a > em')
    info_spec = soup_movie.select_one('dl.info_spec')
    specs = info_spec.select('p')

    star_info = list()
    spc_star = re.sub('(<([^>]+)>)', '', str(spc_star_info))
    ntz_star = re.sub('(<([^>]+)>)', '', str(ntz_star_info))

    if len(spc_star) == 2 :
      spc = '기자, 평론가 평점 없음'
    else : 
      spc = '기자, 평론가 평점 ' + str(star(spc_star))

    if len(ntz_star) == 2 :
      ntz = '네티즌 평점 없음'
    else : 
      ntz = '네티즌 평점 ' + str(star(ntz_star))
      
    star_info.append(main_star_info)
    star_info.append(spc)
    star_info.append(ntz)

    spec_detail = list()
    for spec in specs :
      spec_info =  re.sub('(<([^>]+)>)', '', str(spec))
      spec_info = spec_info.strip()
      spec_info = re.sub('\n\n', '\',\'', spec_info)
      spec_info = re.sub("\s", "", spec_info)
      # spec_info = re.sub('\t\n', '\'', spec_info)
      # spec_info = re.sub('"', '', spec_info)
      spec_detail.append(spec_info)
      a = str(spec_detail[0])
      p = a.split('\',\'')
      for i in range(len(spec_detail)-1):
        p.append(spec_detail[i+1])
      # spec_deatail.

    result_info = title + star_info + p
    print(result_info)
```
